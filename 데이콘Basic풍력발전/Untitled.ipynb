```python
# 라이브러리 불러오기
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings(action='ignore')
%config InlineBackend.figure_format = 'retina'
```

```python
path = 'train.csv'
data_train = pd.read_csv(path)
```

```python
data_train.head()
```

```python
# snowing 부울 값을 숫자로
data_train['snowing'] = data_train['snowing'].replace({False:0,True:1})
data_train.head(3)
```

```python
# 변수확인, 결측치확인, 기술통계확인
display(data_train.info())
print('='*70)
display(data_train.describe().T)
print('='*70)
display(data_train.isna().sum())
print('='*70)
```

```python
#상관관계 시각화 및 확인
plt.figure(figsize=(8,5))
sns.heatmap(data_train.corr(),
           cmap='Blues',
           cbar=False,
           annot=True,
           fmt = '.2f',
           square=True)
```
